##**clustering**

```python
from sklearn.cluster import KMeans

left_employees = df[df['left'] == 1]
features = left_employees[['satisfaction_level', 'last_evaluation']]

num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(features)
left_employees['cluster'] = kmeans.labels_

for i in range(num_clusters):
    cluster_data = left_employees[left_employees['cluster'] == i]
    plt.scatter(cluster_data['satisfaction_level'], cluster_data['last_evaluation'], label=f'Cluster {i}')

plt.title('Clustering of Employees Who Left')
plt.xlabel('Satisfaction Level')
plt.ylabel('Last Evaluation')
plt.legend()
plt.show()
```

__Observation__


The algorithm has grouped the employees who left into three distinct clusters based on their 'satisfaction_level' and 'last_evaluation.' These clusters appear to be well-separated, indicating that employees who left can be categorized into three different groups:



0.   The Blue cluster represents individuals with moderate or also higher satisfaction levels.
1.   
The orange cluster represents individuals with lower satisfaction levels and received scores that are lower compared to the clusters mentioned above.
2.   The Green  cluster represents individuals who have the highest satisfaction levels.





###**Step 4:** Handle the left Class Imbalance using SMOTE technique

- Import train_test_split and split the data
- Import over_sampling to do SMOTE technique

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

# Separate categorical and numeric columns
categorical_cols = df.select_dtypes(include=['object']).columns
numeric_cols = df.select_dtypes(exclude=['object']).columns

# Convert categorical variables to numerical using one-hot encoding
categorical_data = pd.get_dummies(df[categorical_cols], drop_first=True)

# Combine numerical and one-hot encoded categorical data
processed_data = pd.concat([df[numeric_cols], categorical_data], axis=1)

# Assuming 'left' is the target variable and 'X' is the feature matrix
X = processed_data.drop('left', axis=1)
y = processed_data['left']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=123)

smote = SMOTE(random_state=123)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
print("X_train_Resampled",X_train_resampled.shape)
```

##**Step 5:** Perform 5-Fold cross-validation model training and evaluate performance.

- Import models and split the data

```python
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```


*   Let's create and train the logistic regression model.
*   Predict the labels for the test set.
*   And know the accuracy of the model.





```python
lr = LogisticRegression(solver='liblinear')
lr.fit(X_train_resampled, np.ravel(y_train_resampled))
ypred = lr.predict(X_test)

accuracy = lr.score(X_test, y_test)
print ("Accuracy is {}".format(accuracy))
```

```python
print('cross validation is:',cross_val_score(lr, X_train_resampled,y_train_resampled, cv=5).mean())
```

```python
print('classification_report:\n',classification_report(y_test,ypred))
```

*   Let's create and train the Random Forest Classifier model.
*   Predict the labels for the test set.
*   And know the accuracy of the model.

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(max_depth = 5)
rf.fit(X_train_resampled,y_train_resampled)
ypred1=rf.predict(X_test)

accuracy = rf.score(X_test, y_test)
print ("Accuracy is {}".format(accuracy))
```

```python
print('cross validation is:',cross_val_score(rf, X_train_resampled, y_train_resampled, cv=5).mean())
```

```python
print('classification_report:\n',classification_report(y_test,ypred1))
```

*   Let's create and train the Gradient Boosting Classifier  model.
*   Predict the labels for the test set.
*   And know the accuracy of the model.

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(n_estimators=100, learning_rate = 0.1, max_depth=1, random_state=42)
gb.fit(X_train_resampled,y_train_resampled)
ypred2 = gb.predict(X_test)

accuracy = gb.score(X_test, y_test)
print ("Accuracy is {}".format(accuracy))
```

```python
print('cross validation is:',cross_val_score(gb,X_train_resampled , y_train_resampled, cv=5).mean())
```

```python
print('classification_report:\n',classification_report(y_test,ypred2))
```

##**Step 6:** Identify the best model and justify the evaluation metrics used.

```python
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score
import matplotlib.pyplot as plt

models = [lr, rf, gb]
model_names = ["LogisticRegression", "Random Forest", "GradientBoostingClassifier"]

for i, model in enumerate(models):
    proba = model.predict_proba(X_test)
    proba_class1 = proba[:, 1]
    fpr, tpr, thresholds = roc_curve(y_test, proba_class1)
    roc_auc = roc_auc_score(y_test, proba_class1)

    plt.plot(fpr, tpr, label=f'{model_names[i]} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'r--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves for Multiple Models")
plt.legend(loc="lower right")
plt.show()
```

**Observations**
- Here, we can observe the ROC\AUC curve of the models.

```python
print('\nconfusion matrix for LogisticRegression model:\n', confusion_matrix(y_test,ypred))
print('\nconfusion matrix for Random Forest model:\n',confusion_matrix(y_test,ypred1))
print('\nconfusion matrix for GradientBoostingClassifier model:\n',confusion_matrix(y_test,ypred2))
```

**From the confusion matrix, explain which metric needs to be used- Recall or Precision?**

Considering the confusion matrix, it is important to minimize the false negatives. When an employee who is likely to leave the organization is incorrectly classified as someone who will stay, necessary retention strategies may not be implemented for that individual. Therefore, using the recall metric is recommended as it focuses on minimizing false negatives and capturing the true positives.

```python
zone=[]
prob=[]
proba = rf.predict_proba(X_test)

for i in proba[:,1]:
    prob.append(i)
    if i <= 0.2:
        zone.append("Safe Zone")
    elif 0.2 < i <= 0.6:
        zone.append("Low Risk Zone")
    elif 0.6 < i <= 0.9:
        zone.append("Medium Risk Zone")
    else:
        zone.append("High Risk Zone")

c = ["Red","Yellow","Orange","Green"]
clr = pd.DataFrame({"zone":zone,"probability":prob})

# Count the number of employees in each zone
zone_counts = pd.Series(zone).value_counts().sort_index(ascending=True)

# Plot the distribution of employees in different zones
plt.figure(figsize=(7,7))
sns.barplot(x=zone_counts.index, y=zone_counts.values, palette=c)
plt.xlabel('Zone')
plt.ylabel('Number of Employees')
plt.title('Employee Distribution in Different Zones')

plt.show()
```
